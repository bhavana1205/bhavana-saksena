**WHAT IS RANDOM FOREST ALGORITHM? **
^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^
RANDOM forest is a famous machine learing that uses supervise learning methods.
you can apply it to both classification and regression problems.

it is based o esnseble learning,which integrates multiple classifiers to solve a complex issue and increases the model's performance.

in laymans term,random forest is a classifer that contains several decision trees on various subsets of a given dataset and takes the avg to enhance the predicated accuracy of that dataset.
instead of relying on single decision tree the random forest collects the result from each trees and ext the final output based on the majority  votes of perdicator


** working of random forest alg**
^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^
the working of the random forest alg is quite intutive.it implemented in two phases:the first is to combine N decision trees with the building the random forest,and second is to make prediction for each tree created in the first phase.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('kyphosis.csv')
```

```python
df.head()
```

```python
sns.pairplot(df,hue='Kyphosis',palette='Set1')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop('Kyphosis',axis=1)
Y=df['Kyphosis']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree=DecisionTreeClassifier()
```

```python
dtree.fit(X_train,Y_train)
```

```python
predictions =dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(Y_test,predictions) )
```

```python
print(confusion_matrix(Y_test,predictions))
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=100)
rfc.fit(X_train,Y_train)
```

```python
rfc_pred=rfc.predict(X_test)
```

```python
print(confusion_matrix(Y_test,rfc_pred))
```
